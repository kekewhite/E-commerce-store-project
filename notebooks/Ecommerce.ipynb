```python
#importing panda library
import pandas as pd
```

```python
#importing Ecommerce data file as CSV.
#There was an encoding because CSV file is not encoded in UTF-8.
#Showing first 5 rows of the data.

df = pd.read_csv(
    r"C:\Users\USER\Documents\Kwe 4\KWe.2\Ecommerce Data.csv",
    encoding="latin1"
)

df.head()
```

HANDLING MISSING VALUES

```python
#Identify the number of rows in each colmn that has missing values

df.isnull().sum()
```

```python
#CustomerID and description had missing value. So i will remove all missing CustomerID
```

```python
df = df.dropna(subset=['CustomerID'])
```

```python
df.isnull().sum()

#no more missing values
```

```python
FIND AND REMOVE DUPLICATED ROWS
```

```python
#this shows us the rows that duplicates

duplicates = df[df.duplicated()]

duplicates
```

```python
# Removes this duplicate rows

df = df.drop_duplicates()
```

```python
df
```

```python
#I will remove records with quantity </= 0

df = df[df['Quantity'] >= 0]
df
```

```python
df.dtypes
```

CHANGING DATA TYPES

```python
#I Will find invoice no. that are not integers


df.loc[
    ~df['InvoiceNo'].astype(str).str.isdigit(),
    'InvoiceNo'
]
```

```python
#I Will remove the 'C' and replace with nothing

df['InvoiceNo'] = df['InvoiceNo'].astype(str).str.replace('C', '')
```

```python
#I Will convert the invoice no. datatype from object to integers

df['InvoiceNo'] = df['InvoiceNo'].astype(int)
```

```python
#I Will convert the Quantity datatype from object to integers

df['Quantity'] = df['Quantity'].astype(int)
df
```

```python
#I Will convert the nvoiceDate datatype from object to datetime


df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df
```

```python
#I Will convert the CustomerID datatype from floats to integers

df['CustomerID'] = df['CustomerID'].astype(int)
df
```

```python
# I will create a Revenue column by multiplying unitprice and Quantity column

df['Revenue'] = df['Quantity'] * df['UnitPrice']
df
```

```python
# creating a new column with monthname extracted from InvoiceDate

df['Month'] = df['InvoiceDate'].dt.month_name()
df['Month']
```

```python
# creating a new column with monthnumber extracted from InvoiceDate


df['MonthNumber'] = df['InvoiceDate'].dt.month
df['MonthNumber']
```

```python
# creating a new column with Year extracted from InvoiceDate

df['Year'] = df['InvoiceDate'].dt.year
df['Year']
```

```python
df
```

```python
#to get total revenue generated

total_revenue = df['Revenue'].sum()
    
total_revenue


print(f"${total_revenue:,.2f}")
```

```python
#to get total revenue generated in each year

revenue_by_Year = (
    df.groupby(['Year'])['Revenue']
    .sum()
    .reset_index()
)

revenue_by_Year
```

```python
#to get total revenue generated in each month for each year

revenue_by_month = (
    df.groupby(['Year', 'Month'])['Revenue']
    .sum()
    .reset_index()
    .sort_values(by=['Year', 'Month'])
)

revenue_by_month
```

```python
#to get total revenue generated in each month for each year

revenue_by_MonthNumber = (
    df.groupby(['Year', 'MonthNumber'])['Revenue']
    .sum()
    .reset_index()
    .sort_values(by=['Year', 'MonthNumber'])
)

revenue_by_MonthNumber
```

```python
#Top performing customers based on revenue

top_customers = (
    df.groupby('CustomerID')['Revenue']
    .sum()
    .sort_values(ascending=False)
    .head(20)
)

top_customers
```

```python
#Top performing Items based on revenue

top_products = (
    df.groupby('Description')['Revenue']
    .sum()
    .sort_values(ascending=False)
    .head(20)
)

top_products
```

```python
#Top performing Country based on revenue

top_Country = (
    df.groupby('Country')['Revenue']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_Country
```

```python
#AVERAGE REVENUE

df['Revenue'].mean()
```

```python
# USING 23 as mean revenue, lets group customers into spendingtype based on revenue

df['SpenderType'] = df['Revenue'].apply(
    lambda x: 'High Spender' if x > 23 else 'Low Spender'
)
df
```

VISUALIZATION

```python
import matplotlib.pyplot as plt
```

```python
#Create a table showing monthly revenue sorted by Year then month number

revenue_by_month = (
    df.groupby(['Year', 'MonthNumber', 'Month'])['Revenue']
    .sum()
    .reset_index()
    .sort_values(by=['Year', 'MonthNumber'])
)
revenue_by_month
```

```python
# Create a Year-Month column


revenue_by_month['YearMonth'] = (
    revenue_by_month['Month'] + ' ' +
    revenue_by_month['Year'].astype(str)
)
revenue_by_month
```

```python
#CREATE A Line chart

plt.figure(figsize=(12,6))

plt.plot(
    revenue_by_month['YearMonth'],
    revenue_by_month['Revenue'],
    marker='o'
)

plt.title('Revenue by Month')
plt.xlabel('Month')
plt.ylabel('Revenue')

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```

```python
#Top performing Items based on revenue

top_products = (
    df.groupby('Description')['Revenue']
    .sum()
    .sort_values(ascending=False)
    .head(20)
)

top_products
```

```python
# BAR CHART SHOWING TOP PRODUCTS
# Create figure
plt.figure(figsize=(12,6))

# Horizontal bar chart
plt.barh(
    top_products.index,
    top_products.values
)

# Titles and labels
plt.title('Top 10 Products by Revenue')
plt.xlabel('Revenue')
plt.ylabel('Product')

# Highest revenue at top
plt.gca().invert_yaxis()

# Improve spacing
plt.tight_layout()

# Show chart
plt.show()
```

```python
#Top performing customers based on revenue

top_customers = (
    df.groupby('CustomerID')['Revenue']
    .sum()
    .sort_values(ascending=False)
    .head(20)
)

top_customers
```
